```python
create_table_sql = '''
  CREATE TABLE IF NOT EXISTS rainfall (
  id BIGSERIAL PRIMARY KEY,
  ProvinceId INTEGER,
  ProvinceName CHAR(60),
  ProvinceName2 CHAR(60),
  MinRain double precision,
  MaxRain double precision,
  AvgRain double precision,
  region CHAR(20),
  Year CHAR(4),
  Month CHAR(2),
  Date DATE,
  Month_TH CHAR(20)
  );
'''
```

# Insert data to Neon

```python
import pandas as pd
import numpy as np

pd = pd.read_csv("rainfall.csv");
```

```python
import datetime

format = '%d/%m/%y'
```

```python
pd.head()
```

```python
for row in pd.iterrows():
  provinceId = row[1]['ProvinceID']
  provinceName = row[1]['ProvinceName']
  provinceNameTh = row[1]['ProvinceName 2']
  minRain = round(row[1]['MinRain'],3)
  maxRain = round(row[1]['MaxRain'],3)
  avgRain = round(row[1]['AvgRain'],3)
  region = row[1]['region']
  year = row[1]['Year']
  month = row[1]['Month']
  date = datetime.datetime.strptime(row[1]['Date'], format).date()
  monthTh = row[1]['Month_TH']

  insert_statement = "INSERT INTO rainfall (ProvinceId,ProvinceName,ProvinceName2,MinRain,MaxRain,AvgRain,region,Year,Month,Date,Month_TH) VALUES (%d, '%s', '%s', %f, %f, %f, '%s', '%s', '%s', '%s', '%s');" % (provinceId, provinceName, provinceNameTh, minRain, maxRain, avgRain, region, year, month, date, monthTh)
  # Execute the SQL statements
  cursor.execute(insert_statement)
  # Commit the changes
  connection.commit()
```
